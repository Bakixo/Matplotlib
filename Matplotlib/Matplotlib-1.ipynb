```python
x = np.arange(0,10)
y = x**2 - 10*x + 25
```

```python
x,y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(16,3), dpi=200) # Figure metoduyla boyutunu figsize= diyerek ayarlayabiliriz, dpi ise görselin netliğini ayarlar.
plt.plot(x,y)
plt.title("First Graphic", size=15) # grafiğin adı.
plt.xlabel("X ekseni.") # x ekseninin adı.                      "size" yazı boyutunu ayarlar.
plt.ylabel("Y ekseni.",size= 12) # y eksenin adı.
plt.xticks(np.arange(min(x),max(x)+1 ,1)); # aradaki değerleri daha detaylı görmek istersek böyle yapabiliriz (x ekseni için), sondaki 1 kaçar kaçar gitmesi gerektiğiyle alakalı.
plt.yticks(np.arange(min(y),max(y)+1, 2.5)); # bu da aynı şey sadece y ekseni için. SONDAKİ ";" kodu çalıştırdığımızda üstünde yazan yazı kalabalığı için.
```

```python
plt.figure(figsize=(16,5))
plt.plot(x,y)
plt.xticks(np.arange(min(x),max(x)+ 1,1));
plt.yticks(np.arange(min(y),max(y), 2.5));

plt.xlim(3,5) # sadece x eksenin de 3 den 7 ye kadar olan yeri çizmesini istedik.
plt.ylim(0,5) # bu da sadece 0 ile 5 arasındaki y eksenini aldı.


# plt.show() üstüne yazdığınız tüm kodu sonuna ; koyduğumuz şekilde yansıtır!!!
```

Subplotların arasındaki boşlukları isteğe göre düzenlemek için: https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots_adjust.html

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1][0].plot(x,x**2)
ax[1,1].plot(x,x**0.5)

fig.subplots_adjust(
    left=None,
    bottom=None, # None boşukları kaldırmamız için
    right=None,
    top=None,
    wspace=0.6,
    hspace=0.6,
    )
# plt.tight_layout() ---> Verimsiz olduğu durumda kullanılabilir.
```

```python
fig, ax = plt.subplots(2,2)

ax[0][0].plot(x,x)
ax[0,1].plot(x,x*2)
ax[1][0].plot(x,x**2)
ax[1,1].plot(x,x**0.5)

plt.tight_layout();

# Kaydetmek için
fig.savefig('figure_2.png',bbox_inches='tight') # bbox_inches indirilidiğinde çıkan fazlalıkları kesti
```

# Eksen Düzenlemek

### Logaritmik eksen

```python
fig, ax = plt.subplots(1,2)

ax[0].plot(x,np.exp(x))
ax[0].set_title("Normal Grafik")

ax[1].plot(x,np.exp(x))
ax[1].set_yscale("log")
ax[1].set_title("Logaritmik eksen (y)")

plt.tight_layout();
```

### İki Eksenli Grafik

```python
fig, ax1 = plt.subplots()

ax1.plot(x, 4*x, lw=2, color="blue", marker="o")
ax1.set_ylabel(r"Çevre $(4X)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx() #twinx çift eksen olması için hem sağda hem solda.
ax2.plot(x, x**2, lw=2, color="red", marker="s")
ax2.set_ylabel(r"Alan $(X^2)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")

```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none') # sağdaki çizgiyi kaldır.
ax.spines['top'].set_color('none') # üstteki çizgiyi kaldır.

ax.xaxis.set_ticks_position('bottom') # x eksenine ortaya bir çizgi koy.
ax.spines['bottom'].set_position(('data',0)) # 0 noktasında bir çizgi çekmek istiyorum.

ax.yaxis.set_ticks_position('left') 
ax.spines['left'].set_position(('data',0))

x = np.linspace(-4, 3, 100)
y = 1.7*x**3 + 3.7*x**2 - 2.36*x + 7.62
ax.plot(x, y);
```

```python
fig, ax = plt.subplots()

plt.axvline(0, c='grey', ls="--", lw=1)
plt.axhline(0, c='grey', ls="--", lw=1)

ax.plot(x, x**2, x, np.log(x))

ax.text(-3, 9.5, r"$y=x^2$", fontsize=12, color="blue")
ax.text(0.3, -3, r"$y=log(x)$", fontsize=12, color="green")
```

## 3D Grafik

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
```

```python
from matplotlib import cm
from matplotlib.ticker import LinearLocator

fig, ax = plt.subplots(figsize=(7,5), dpi=200, subplot_kw={"projection": "3d"})
# Make data.
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm,
                       linewidth=0, antialiased=False)

# Customize the z axis.
ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
